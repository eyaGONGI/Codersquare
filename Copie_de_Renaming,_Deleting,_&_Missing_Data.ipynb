```python
import pandas as pd
bike_df=pd.read_excel('/content/RailsToTrails_National Count Data_week (1).xlsx')
bike_df.head(5)
```

```python
bike_df_dropped=bike_df.drop(columns=['Unnamed: 7'	, 'Unnamed: 8'	, 'Unnamed: 9'])
bike_df_dropped.head(5)
```

```python
bike_df_dropped= bike_df_dropped.rename(columns={'Unnamed: 6' : 'percent_change'})
bike_df_dropped.head(5)
```

How many missing values?

```python
bike_df_dropped.isna().sum()
```

Filling missing values

```python
bike_df_dropped['2021 Counts'] = bike_df_dropped['2021 Counts'].fillna(0)
```

```python
bike_df_dropped.head()
```

```python
bike_df_dropped= bike_df_dropped.rename(columns={ ' 2019 counts (31 counters)' : 'counts_2019'})
```
